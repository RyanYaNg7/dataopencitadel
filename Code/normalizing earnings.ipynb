```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

earnings = pd.read_csv("earnings_updated_fips.csv")

earnings.head()

```

```python
earnings.fillna(0, inplace=True)
earnings.replace(['2,500-', '250,000+'], [1500, 300000], inplace = True)
```

```python
earnings.set_index(earnings.columns.values[0], inplace = True)
earnings.set_index(earnings.columns.values[0], inplace = True)
earnings.set_index(earnings.columns.values[0], inplace = True)
earnings
```

```python
selected_earnings = earnings.drop(columns = ['county', 'total_med', 'year'])
s = selected_earnings.astype('float') 
s
```
