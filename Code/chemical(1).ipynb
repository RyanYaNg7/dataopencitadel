```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

chem_data = pd.read_csv("chemicals.csv")

chem_data.head()
```

```python
chem_Uran = chem_data[chem_data.chemical_species == "Uranium"]
chem_Ars = chem_data[chem_data.chemical_species == "Arsenic"]
chem_DEHP = chem_data[chem_data.chemical_species == "DEHP"]
chem_Nitrates = chem_data[chem_data.chemical_species == "Nitrates"]
chem_Halo = chem_data[chem_data.chemical_species == "Halo-Acetic Acid"]
chem_Tri = chem_data[chem_data.chemical_species == "Trihalomethane"]
```

```python
# get new dataframe with aggregate data for each year

counties = chem_data.fips.unique()
years = chem_data.year.unique()
chemicals = chem_data.chemical_species.unique()
# years = [str(x) for x in years]
columns = years

chem_data_Uran = pd.DataFrame(0, index = counties, columns = columns) 
chem_data_Ars = pd.DataFrame(0,index = counties, columns = columns)
chem_data_DEHP = pd.DataFrame(0,index = counties, columns = columns)
chem_data_Nitrates = pd.DataFrame(0,index = counties, columns = columns)
chem_data_Halo = pd.DataFrame(0,index = counties, columns = columns)
chem_data_Tri = pd.DataFrame(0,index = counties, columns = columns)

chem_data_Total = pd.DataFrame(0,index = counties, columns = columns)

#chem_data_pop = pd.DataFrame(0,index = counties, columns = columns)


measures = chem_data.unit_measurement.unique()

chem_dict = {}
for i, chem in enumerate(chemicals):
    chem_dict[chem]  = i

chem_dict
```

```python
from __future__ import division

current_county = 'Alameda'
county_pop = 0
county_tot = 0


for row in chem_Uran.itertuples():
    year = row[10]
    val = row[9]
    pop = row[6]
    county = row[11]
    chemical = row[2]
    measure = row[8]
    if county == current_county: 
        county_pop += pop
        if measure == measures[1] and county_pop != 0:
            county_tot += val*pop*1000
            chem_data_Uran.loc[county][year] = county_tot*100/county_pop 
        elif county_pop != 0:
            county_tot += val*pop
            chem_data_Uran.loc[county][year]= county_tot*100/county_pop
    else:
        current_county = county
        county_pop = pop
        county_tot = 0
        if measure == measures[1] and county_pop != 0:
            county_tot += val*pop*1000
            chem_data_Uran.loc[county][year] = county_tot*100/county_pop 
        elif county_pop != 0:
            county_tot += val*pop
            chem_data_Uran.loc[county][year] = county_tot*100/county_pop
        
        
    
    

```
